```python
NewDf = MyDf 
NewDf.head()
```

```python
# I am taking out the columns I am not immediately concerned with
MyDf.drop(['iso_code','new_cases','new_cases_smoothed','new_deaths','new_deaths_smoothed',
                   'total_cases_per_million','new_cases_per_million','new_cases_smoothed_per_million',
                  'total_deaths_per_million','new_deaths_per_million','new_deaths_smoothed_per_million',
                  'icu_patients_per_million','hosp_patients','hosp_patients_per_million','weekly_icu_admissions',
                  'weekly_icu_admissions_per_million','weekly_hosp_admissions','weekly_hosp_admissions_per_million',
                  'new_tests','total_tests_per_thousand','new_tests_per_thousand','new_tests_smoothed',
                  'new_tests_smoothed_per_thousand','positive_rate','tests_per_case','tests_units',
                  'total_vaccinations_per_hundred','stringency_index','population','population_density',
                  'median_age','gdp_per_capita','extreme_poverty','cardiovasc_death_rate','diabetes_prevalence',
                  'female_smokers','male_smokers','handwashing_facilities','hospital_beds_per_thousand'],axis=1,inplace=True)
```

```python
NewDf = NewDf.dropna(how='all')
NewDf.head()
```

```python
# This is not necessary. I only do it to make sure all fields that should be numeric are actually numeric
NewDf['total_cases'] = pd.to_numeric(NewDf['total_cases'])
NewDf['total_deaths'] = pd.to_numeric(NewDf['total_deaths'])
NewDf['reproduction_rate'] = pd.to_numeric(NewDf['reproduction_rate'])
NewDf['icu_patients'] = pd.to_numeric(NewDf['icu_patients'])
NewDf['total_tests'] = pd.to_numeric(NewDf['total_tests'])
NewDf['total_vaccinations'] = pd.to_numeric(NewDf['total_vaccinations'])
NewDf['aged_65_older'] = pd.to_numeric(NewDf['aged_65_older'])
NewDf['life_expectancy'] = pd.to_numeric(NewDf['life_expectancy'])
NewDf['human_development_index'] = pd.to_numeric(NewDf['human_development_index'])

```

```python
NewDf
```

```python
# Making sure the date field is recognised as date 
NewDf['date'] = NewDf['date'].map(pd.to_datetime)
NewDf.head()
```

```python
# Get actual Day and month on new columns

# for actual day
def get_dom(dt):
    return dt.day
NewDf["Actual Day"] = NewDf["date"].map(get_dom)

# for actual month
def get_moy(mt):
    return mt.month
NewDf["Actual Month"] = NewDf["date"].map(get_moy)


NewDf.head()
```

```python
# Active cases
# Subtract 'Total cases' from the 'total deaths' and 'recoveries'.
# Since we don't have recovery column, we just use what we have

NewDf['Active_cases']=NewDf['total_cases']-NewDf['total_deaths']
NewDf
```

```python
NewDf.groupby('location')['Active_cases'].sum().sort_values(ascending=False).to_frame()
```

```python
# Delete record for world
NewDf = NewDf.set_index("location")
NewDf = NewDf.drop("World", axis=0)
```

```python
# Totaal Active Cases
totalActiveCases=NewDf.groupby('location')['Active_cases'].sum().sort_values(ascending=False).to_frame()
```

```python
totalActiveCases.style.background_gradient(cmap='Blues')
```

```python
# Total unfortunate deaths
totalDeaths=NewDf.groupby('location')['total_deaths'].sum().sort_values(ascending=False).to_frame()
totalDeaths.style.background_gradient(cmap='Reds')
```

```python
# Days with the highest activities in each month.
# This is to demonstrate how to look at all the numeric columns at the same time
# Look at the columns you're concerned with and ignore others

hotestDay = NewDf.groupby('Actual Day').sum()
hotestDay.style.background_gradient(cmap="Greens")
```

```python
# Days of the month with the highest cases

hotDays=NewDf.groupby('Actual Day')['total_cases'].sum().sort_values(ascending=False).to_frame()
hotDays.style.background_gradient(cmap='Blues')
```

```python
# Days of the month with the highest unfortunate deaths

redDays=NewDf.groupby('Actual Day')['total_deaths'].sum().sort_values(ascending=False).to_frame()
redDays.style.background_gradient(cmap='Reds')
```

```python
# Months with the highest activities.

hotestMonths = NewDf.groupby('Actual Month').sum()
hotestMonths.style.background_gradient(cmap='Greens')
```
